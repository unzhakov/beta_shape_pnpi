```python
from beta_spectrum.components.phase_space import PhaseSpace
from beta_spectrum.components.fermi import FermiFunction
from beta_spectrum.utils import T_to_W
from numpy import arange, trapezoid

import matplotlib.pyplot as plt
```

```python
endpoint_MeV = 0.2953
e_step_MeV = 0.001
Z_d = 44
Z_p = 43
A = 99

W0 = T_to_W(endpoint_MeV)
E = arange(e_step_MeV, endpoint_MeV, e_step_MeV)
W = T_to_W(E)
tr_type = "F2"
```

```python
ff = FermiFunction(Z_p, A)
```

```python
for forb in ["F1", "F2","F3", "F4"]:
    ph = PhaseSpace(W0, transition_type=forb)
    integral = trapezoid(ph(W), W)
    plt.plot(W, ff(W) * ph(W)/integral, label=forb)
    # plt.yscale("log")
    plt.legend()
```
