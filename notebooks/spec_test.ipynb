```python
from beta_spectrum.spectrum import SpectrumConfig, BetaSpectrum, BetaSpectrumAnalyzer
```

```python
# Tc-99
config = SpectrumConfig(
    Z_parent=43,
    Z_daughter=44,
    A_number=99,
    endpoint_MeV=0.2935,
    e_step_MeV=0.0001,  # 100 eV
    # use_exchange=False,
)
```

```python
spectrum = BetaSpectrum.from_config(config)
analyzer = BetaSpectrumAnalyzer(spectrum, config)
```

```python
analyzer.plot_analysis(save_path='tc99_plots.png')
```

```python
analyzer.export_to_csv('tc99_spectrum.csv')
```
